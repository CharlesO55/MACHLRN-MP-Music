# 1.2 Fixing the missing values.

There are 3 options available.
<ol>
    <li>Ignore the missing values.</li>
    <li>Remove the missing rows.</li>
    <li>Impute/Replace the missing values.</li>
</ol>

Ignoring the issue may cause accuracy problems while removing the missing rows would wipe 6,183 of 17,996 rows (__34% of all the data__). 

The ideal solution is to choose imputation.<br>
This method allows us to replace missing values with estimates instead.



### First let's check if there's enough data points to make estimations.
We will be choosing to make our estimates grouped by the Class (genre) type.<br>
This allows the estimates to be localized within the genre instead of generalizing with a global estimate.

```python
dataLoader.dict_df['raw']['Class'].value_counts()
```

### Although severely imbalanced, it is possible to make estimates with at least 300+ entries.

### Next, let's decide on an estimation/aggregation method.

```python
dataLoader.dict_df['raw'][['key', 'instrumentalness']].boxplot()
```

### Using Median 
It's seen in the boxplot that our values contain numerous outliers skewed entirely in one direction.<br>
Median is less affected by these outliers hence it will be the chosen method.

### The median estimates generated:

```python
dataLoader.dict_df['raw'].groupby('Class')[['key', 'instrumentalness']].median()
```

## Everything done was performed in DataLoader.do_imputation and stored to `dict_df['imputed']`
'''

    def do_Imputation(self):
    
        # Find the rows with missing values
        mask = self.dict_df['raw'].isna().any(axis=1)
    
        # Generate the median estimates
        self.estimates = self.dict_df['raw'].groupby('Class')[['key', 'instrumentalness']].median()
    
        # Create a copy to work on
        self.dict_df['imputed'] = self.dict_df['raw'].copy()
        
        # Replace all NaN values under `key` and `instrumentalness`
        self.dict_df['imputed'] = self.dict_df['imputed'].apply(self.impute_Rows, axis=1)
    
    def is_NaN(self, x):
    
        return x != x;
        
    def impute_Rows(self, X, cols_to_Impute = ['key', 'instrumentalness']):
    
        for col in cols_to_Impute:
            if self.is_NaN(X[col]):
                X[col] = self.estimates.loc[X['Class'], col]
      
        return X
'''

```python
dataLoader.do_Imputation()
```

## NaN have now been imputed/replaced.

# 1.3 Feature Engineering
Let's proceed to feature engineering.
#### Consider the features that can be further extracted or combined.

### Track Names
Although `Track Name` has corrupted values, it still possible to extract additional details such as:
<ol>
    <li>Is_Collab {Member(s) only = 0, External collab = 1 } - If it contains the keywords ` ft.`, `feat` or `Feat.`</li>
    <li>Is_Remix {Original = 0, Remix = 1} - If it contains the keywords `remix` or `Remix`</li>
</ol>

### `DataLoader.do_Feature_Engineering_Track_Names` does this and stores to `dict_df['feature_eng']`
'''

    def find_tracks_with_string(df, text_to_search):
        return df.query("`Track Name`.str.contains(@text_to_search)").index
    
    # CREATE NEW COLS
    df[['Is_Collab', 'Is_Remix']] = np.zeros((df.shape[0], 2))
    
    # ACTIVATE THE FLAGS 
    df.loc[find_tracks_with_string('feat'), 'Is_Collab'] = 1
    df.loc[find_tracks_with_string('Feat'), 'Is_Collab'] = 1
    df.loc[find_tracks_with_string('remix'), 'Is_Remix'] = 1
    df.loc[find_tracks_with_string('Remix'), 'Is_Remix'] = 1
    
    df.query("`Is_Remix` == 1 | `Is_Collab` == 1")
'''

```python
dataLoader.do_Feature_Engineering_Track_Names()
```

## Genres performed
In this sense, we will be keeping logs of the genres each artist performed.

This gives us binaries of `HasSung_(Class no)` where 1 means they've perfommed that genre before.

To do this, We aggregate by the Artist and select the unique values from their `Class` column.

```python
df_artists = dataLoader.dict_df['feature_eng'][['Artist Name', 'Class']].groupby('Artist Name')['Class'].unique()
df_artists
```

### Create new cols containing the HasSung_(class No) with default 0 value

```python
df = dataLoader.dict_df['feature_eng'].copy()

has_Sung_Cols = []
for i in range(0, 11):
    has_Sung_Cols.append('HasSung_' + str(i))

df.loc[:, has_Sung_Cols] = 0
df
```

### Go through each row and flag the genres they've sung.

```python
def flag_Sung_Genres(row):
    artist = row['Artist Name']

    for genres in df_artists[artist]:
        row['HasSung_' + str(genres)] = 1

    return row

# df.apply(flag_Sung_Genres, axis=1)
```
